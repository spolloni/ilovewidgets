```python
from sympy import symbols, lambdify, Piecewise
import numpy as np

import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import (
    BoundedFloatText, Layout, Label, Button,
    Output, GridBox, Tab, VBox, HBox
)
import pandas as pd 
from IPython.display import clear_output, display
plot_output = Output()
table_output = Output()

all_output = HBox(children=[plot_output, table_output])

# labels
label_plan_1 = Label("Basic", layout=Layout(width='auto', grid_area='lp1'))
label_plan_2 = Label("Pro", layout=Layout(width='auto', grid_area='lp2'))
label_plan_3 = Label("Deluxe", layout=Layout(width='auto', grid_area='lp3'))
label_plan_4 = Label("Primo", layout=Layout(width='auto', grid_area='lp4'))
label_mopric = Label("Monthly Price:", layout=Layout(width='auto', grid_area='lmp'))
label_mostok = Label("Monthly Unlocks:", layout=Layout(width='auto', grid_area='lms'))
label_ovpric = Label("Overage Price:", layout=Layout(width='auto', grid_area='lop'))
label_nusers = Label("Number of Users:", layout=Layout(width='auto', grid_area='nus'))

# inputs
monthly_price_1 = BoundedFloatText(value=50,
                                   min=0,
                                   max=1000.0,
                                   step=1,
                                   layout=Layout(width='auto', grid_area='mp1'),
                                   disabled=False,)
monthly_price_2 = BoundedFloatText(value=150,
                                   min=0,
                                   max=1000.0,
                                   step=1,
                                   layout=Layout(width='auto', grid_area='mp2'),
                                   disabed=False,)
monthly_price_3 = BoundedFloatText(value=300,
                                   min=0,
                                   max=1000.0,
                                   step=1,
                                   layout=Layout(width='auto', grid_area='mp3'),
                                   disabled=False,)
monthly_price_4 = BoundedFloatText(value=550,
                                   min=0,
                                   max=1000.0,
                                   step=1,
                                   layout=Layout(width='auto', grid_area='mp4'),
                                   disabled=False,)
monthly_stock_1 = BoundedFloatText(value=25,
                                   min=0,
                                   max=1000.0,
                                   step=1,
                                   layout=Layout(width='auto', grid_area='ms1'),
                                   disabled=False,)
monthly_stock_2 = BoundedFloatText(value=100,
                                   min=0,
                                   max=1000,
                                   step=1,
                                   layout=Layout(width='auto', grid_area='ms2'),
                                   disabled=False,)
monthly_stock_3 = BoundedFloatText(value=250,
                                   min=0,
                                   max=1000.0,
                                   step=1,
                                   layout=Layout(width='auto', grid_area='ms3'),
                                   disabled=False,)
monthly_stock_4 = BoundedFloatText(value=500,
                                   min=0,
                                   max=1000.0,
                                   step=1,
                                   layout=Layout(width='auto', grid_area='ms4'),
                                   disabled=False,)
overage_price_1 = BoundedFloatText(value=4,
                                   min=0,
                                   max=10.0,
                                   step=0.1,
                                   layout=Layout(width='auto', grid_area='op1'),
                                   disabled=False,)
overage_price_2 = BoundedFloatText(value=2.50,
                                   min=0,
                                   max=10.0,
                                   step=0.1,
                                   layout=Layout(width='auto', grid_area='op2'),
                                   disabled=False,)
overage_price_3 = BoundedFloatText(value=1.50,
                                   min=0,
                                   max=10.0,
                                   step=0.1,
                                   layout=Layout(width='auto', grid_area='op3'),
                                   disabled=False,)
overage_price_4 = BoundedFloatText(value=1,
                                   min=0,
                                   max=10.0,
                                   step=0.1,
                                   layout=Layout(width='auto', grid_area='op4'),
                                   disabled=False,)
number_users_1 = BoundedFloatText(value=500,
                                  min=0,
                                  max=10000.0,
                                  step=20,
                                  layout=Layout(width='auto', grid_area='nu1'),
                                  disabled=False,)
number_users_2 = BoundedFloatText(value=200,
                                  min=0,
                                  max=10000.0,
                                  step=20,
                                  layout=Layout(width='auto', grid_area='nu2'),
                                  disabled=False,)
number_users_3 = BoundedFloatText(value=100,
                                  min=0,
                                  max=10000.0,
                                  step=20,
                                  layout=Layout(width='auto', grid_area='nu3'),
                                  disabled=False,)
number_users_4 = BoundedFloatText(value=25,
                                  min=0,
                                  max=10000.0,
                                  step=20,
                                  layout=Layout(width='auto', grid_area='nu4'),
                                  disabled=False,)
# button
model_button = Button(description='Model',
                      button_style='info', # 'success', 'info', 'warning', 'danger' or ''
                      tooltip='Click me',
                      icon='fa-area-chart',
                      layout=Layout(width='auto', grid_area='cb'),
                      disabled=False,)


# tabs & full input widget
tab2_widgets = [
                     label_plan_1,    label_plan_2,
    label_mopric, monthly_price_1, monthly_price_2,
    label_mostok, monthly_stock_1, monthly_stock_2,
    label_ovpric, overage_price_1, overage_price_2,
    label_nusers, number_users_1,  number_users_2,
]
tab3_widgets = [
                     label_plan_1,   label_plan_2,     label_plan_3,
    label_mopric, monthly_price_1, monthly_price_2, monthly_price_3,
    label_mostok, monthly_stock_1, monthly_stock_2, monthly_stock_3,
    label_ovpric, overage_price_1, overage_price_2, overage_price_3,
    label_nusers, number_users_1,  number_users_2,  number_users_3, 
]
tab4_widgets = [
                     label_plan_1,   label_plan_2,     label_plan_3,    label_plan_4,
    label_mopric, monthly_price_1, monthly_price_2, monthly_price_3, monthly_price_4,
    label_mostok, monthly_stock_1, monthly_stock_2, monthly_stock_3, monthly_stock_4,
    label_ovpric, overage_price_1, overage_price_2, overage_price_3, overage_price_4,
    label_nusers, number_users_1,  number_users_2,  number_users_3,  number_users_4
]
tab2 = GridBox(children=tab2_widgets,
        layout=widgets.Layout(
            width='100%',
            grid_template_rows='auto auto auto',
            grid_template_columns=' 15% 15% 15%',
            grid_template_areas='''
            "  . lp1 lp2"
            "lmp mp1 mp2"
            "lms ms1 ms2"
            "lop op1 op2"
            "nus nu1 nu2"
            '''))
tab3 = GridBox(children=tab3_widgets,
        layout=widgets.Layout(
            width='100%',
            grid_template_rows='auto auto auto auto',
            grid_template_columns=' 15% 15% 15% 15%',
            grid_template_areas='''
            "  . lp1 lp2 lp3"
            "lmp mp1 mp2 mp3"
            "lms ms1 ms2 ms3"
            "lop op1 op2 op3"
            "nus nu1 nu2 nu3"
            '''))
tab4 = GridBox(children=tab4_widgets,
        layout=widgets.Layout(
            width='100%',
            grid_template_rows='auto auto auto auto auto',
            grid_template_columns=' 15% 15% 15% 15% 15%',
            grid_template_areas='''
            "  . lp1 lp2 lp3 lp4"
            "lmp mp1 mp2 mp3 mp4"
            "lms ms1 ms2 ms3 ms4"
            "lop op1 op2 op3 op4"
            "nus nu1 nu2 nu3 nu4"
            '''))
tab = Tab(children=[tab2, tab3, tab4], selected_index=1)
tab.set_title(0, 'two plans')
tab.set_title(1, 'three plans')
tab.set_title(2, 'four plans')
input_board = VBox(children=[tab, model_button])


def monthly_cost(price, qty, overage, xmin, xmax):
    if overage <= 0 : overage = 0.00001
    x = symbols('x')
    f_x = Piecewise((price, x <= qty),(price + overage*(x-qty), True))
    lam_x = lambdify(x, f_x, modules=['numpy'])
    x_vals = np.linspace(xmin, xmax, 100)
    y_vals = lam_x(x_vals)
    return x_vals, y_vals 


def ensure_consistency(monthly_prices, monthly_stocks, overage_prices):
    messages = []
    if not all(i < j for i, j in zip(monthly_prices, monthly_prices[1:])):
        messages.append("\n Greater plans need to be more expensive per month.")
    if not all(i < j for i, j in zip(monthly_stocks, monthly_stocks[1:])):
        messages.append("\n Greater plans need to have more unlocks per month.")
    if not all(i > j for i, j in zip(overage_prices, overage_prices[1:])):
        messages.append("\n Smaller plans need to have more expensive overages.")
    return messages
        
    
def model(b):
    
    # collect objects
    plans = tab.selected_index + 2
    
    monthly_prices = [
        monthly_price_1.value,
        monthly_price_2.value,
        monthly_price_3.value,
        monthly_price_4.value,
    ]
    
    monthly_stocks = [
        monthly_stock_1.value,
        monthly_stock_2.value,
        monthly_stock_3.value,
        monthly_stock_4.value,
    ]
    
    overage_prices = [
        overage_price_1.value,
        overage_price_2.value,
        overage_price_3.value,
        overage_price_4.value,
    ]
    
    number_users = [
        number_users_1.value,
        number_users_2.value,
        number_users_3.value,
        number_users_4.value,
    ]
    
    label_plans  = [
        label_plan_1, label_plan_2, 
        label_plan_3, label_plan_4
    ]
    
    pricing_schedule = zip(monthly_prices,
                           monthly_stocks,
                           overage_prices)
    
    messages = ensure_consistency(monthly_prices[:plans],
                                  monthly_stocks[:plans],
                                  overage_prices[:plans])
    
    if messages: 
        plot_output.clear_output(True)
        table_output.clear_output(True)
        with plot_output:
            print("Inconsistencies found in inputted plans:" + "".join(messages))
    else:
        
        # plot x-bounds 
        xmin = round(monthly_stocks[0]*0.3)-1
        xmax = round(monthly_stocks[plans-1]*1.1 +1)
        
        # objects to be initialized
        optimal_choice = None
        y_vals_previous = None
        rownames = [
            "regular price per unlock", 
            "overage price per unlock", 
            "total monthly revenue"]
        summarytable = pd.DataFrame(None, index=rownames)
        
        
        with plt.style.context(('seaborn')):
            f, ax = plt.subplots(figsize=(8,6))
            for i, j in enumerate(pricing_schedule):
                if i < plans:
                    
                    # plot each plan iteratively
                    x_vals, y_vals = monthly_cost(*j, xmin, xmax)
                    ax.plot(x_vals, 
                            y_vals, 
                            linewidth=3, 
                            zorder=10+plans, 
                            label=label_plans[i].value)
                    
                    # update minimum cost curve
                    if optimal_choice is None:
                        optimal_choice = np.inf*np.ones(y_vals.shape)
                    optimal_choice = np.vstack([optimal_choice, y_vals]).min(0)
                
                    # identify indifference points
                    if y_vals_previous is not None:
                        idx = (np.abs(y_vals - y_vals_previous)).argmin()
                        plt.axvline(x=x_vals[idx], color='0', alpha=0.2, ls='--', lw=1)
                        #plt.text(10.1,0,'blah',rotation=90)
                    y_vals_previous = y_vals
                    
                    # add column to summary table
                    info_col = [
                        '${:,.2f}'.format(monthly_prices[i]/monthly_stocks[i]),
                        '${:,.2f}'.format(overage_prices[i]),
                        '${:,.0f}k'.format(number_users[i]*monthly_prices[i]/1000)
                    ]
                    info_col = pd.DataFrame(info_col , columns=[label_plans[i].value], index=rownames)
                    summarytable = pd.concat([summarytable, info_col], axis=1)
                    
                    #print(lol)
            
            # plot minimum cost curve
            ax.plot(x_vals, 
                    optimal_choice, 
                    color='y',
                    alpha=0.3, 
                    linewidth=10,
                    zorder=1,
                    label='Cheapest plan')
            
            # labels and style 
            plt.ylabel("Monthly Cost ($)", fontsize=14)
            plt.xlabel("Number of Unlocks", fontsize=14)
            plt.ylim(bottom=int(0.2*monthly_price_1.value),
                     top=int(1.3*y_vals_previous[-1]))
            plt.legend()
            plt.tight_layout()
            plt.close()
            ax.margins(0.02)
            
            # refresh output
            plot_output.clear_output(True)
            with plot_output:
                display(ax.figure)
                
            table_output.clear_output(True)
            with table_output:
                display(summarytable)
            

display(input_board)
model_button.on_click(model)
display(all_output)
```
